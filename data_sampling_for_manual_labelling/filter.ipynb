```python
import pandas as pd
file_name = "./out_20200414_105818/combined.csv"
df = pd.read_csv(file_name, encoding='utf8', dtype=str)
df.head()
```

```python
# below is the sentiment analysis code rewritten for sentence-level analysis
# note the new module -- word_tokenize!
import nltk.data
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from nltk import sentiment
from nltk import word_tokenize
import numpy as np

# Next, we initialize VADER so we can use it within our Python script
sid = SentimentIntensityAnalyzer()

# We will also initialize our 'english.pickle' function and give it a short name

tokenizer = nltk.data.load('tokenizers/punkt/english.pickle')

for index, row in df.iterrows():
    message_text = row['text']
    # The tokenize method breaks up the paragraph into a list of strings. In this example, note that the tokenizer is confused by the absence of spaces after periods and actually fails to break up sentences in two instances. How might you fix that?
    if(pd.isnull(message_text)):
        continue
    #print(index)
    sentences = tokenizer.tokenize(message_text)

    # We add the additional step of iterating through the list of sentences and calculating and printing polarity scores for each one.

    for sentence in sentences:
            #print(sentence)
            scores = sid.polarity_scores(sentence)
            if(scores['pos'] >= 0.3 or scores['neg'] >= 0.3):
                print(index,"has emotion")
                print()
#             for key in sorted(scores):
#                     print('{0}: {1}, '.format(key, scores[key]), end='')
            #print()
            #print()
    
    if(index==100):
        break
```

```python
def checksentiment(row):
    message_text = row['text']
    if(pd.isnull(message_text)):
        row['with_sentiment'] = 'Error'
        return row
    sentences = tokenizer.tokenize(message_text)
    for sentence in sentences:
        scores = sid.polarity_scores(sentence)
        if(scores['pos'] >= 0.3 or scores['neg'] >= 0.3):
            row['with_sentiment'] = 'True'
  
    return row


def filterwithoutsentiment(df):
    import time
    time_start = time.time()
    print("  -> filterwithoutsentiment()", end='')
    
    df['with_sentiment'] = 'False'
    
    df = df.apply(checksentiment, axis=1)
    
    #remove = df.with_sentiment.isin(['False'])
    #result = df[~remove]
    result = df
    
    processed = len(df.index)
    skipped = processed - len(result.index)
    skipped_percentage = skipped / processed * 100
    time = time.time() - time_start
    print(" - Processed: {:,} | Non-Eng Skipped: {:,} ({:.2f}%) | Time: {:,.3f} sec".format(processed, skipped, skipped_percentage, time))
    
    return result
```

```python
result = filterwithoutsentiment(df)
result.head(20)
```

```python
result.to_csv('new_sentiment.csv', mode='w', header=True, index=False, encoding='utf-8')
```

```python
from langdetect import detect_langs

def checklang(row):
    try:
        re = detect_langs(row['text'])
    except:
        row['is_eng'] = 'False'
    else:    
        if(re[0].lang == "en"):
            row['is_eng'] = 'True'
        else:
            row['is_eng'] = 'False'
  
    return row


def filtereng(df):
    import time
    time_start = time.time()
    print("  -> filter_eng()", end='')
    
    df['is_eng'] = 'False'
    
    df = df.apply(checklang, axis=1)
    
    #remove = df.with_sentiment.isin(['False'])
    #result = df[~remove]
    result = df
    
    processed = len(df.index)
    skipped = processed - len(result.index)
    skipped_percentage = skipped / processed * 100
    time_used = time.time() - time_start
    print(" - Processed: {:,} | Non-Eng Skipped: {:,} ({:.2f}%) | Time: {:,.3f} sec".format(processed, skipped, skipped_percentage, time_used))
    
    return result
```

```python
result = filtereng(df)
result.head()
```

```python
result.to_csv('onlyengandsentiment.csv', mode='w', header=True, index=False, encoding='utf-8')
```
