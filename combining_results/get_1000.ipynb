```python
import pandas as pd
from glob import glob
import os
```

```python
with open('lexicon.txt', 'r') as f:
    x = f.read().splitlines()
    
x = list(map(lambda x:x.lower(),x))
```

```python
def check(text):
    text = str(text)
    text = text.lower()
    for i in x:
        if(text.find(i) != -1):
            return True
    return False

# df['relevant'] = df['text'].apply(check)

# df = df[df['relevant'] == True]
# df
```

```python
# subsets = [file for file in glob('./corona/out_*_sentiment.csv'.format())]
file1 = 'new_combined_sentiment.csv'
file2 = 'combined_with_original_tweet_sentiment.csv'
subsets = [file1,file2]
subsets = sorted(subsets)
print("List of csv: {}\n".format(subsets))
print(len(subsets))
```

```python
for filename in subsets:
    df = pd.read_csv(filename, encoding='utf8', dtype=str, escapechar='\\',error_bad_lines=False,engine='python')
    df['relevant'] = df['text'].apply(check)
    df = df[df['relevant'] == True]
    out_file_name = '23_sentiment_date.csv'
    if not os.path.exists(out_file_name):
        df.to_csv(out_file_name, mode='w', header=True, index=False, encoding='utf-8')
    else:
        df.to_csv(out_file_name, mode='a', header=False, index=False, encoding='utf-8')
```
