```python
import pandas as pd
from glob import glob
import os
```

```python
subsets = [file for file in glob('./corona/out_*_sentiment_date.csv'.format())]
subsets = sorted(subsets)
print("List of csv: {}\n".format(subsets))
print(len(subsets))
```

```python
for filename in subsets:
    df = pd.read_csv(filename, encoding='utf8', dtype=str, escapechar='\\',error_bad_lines=False,engine='python')
    out_file_name = './corona/april_sentiment_date.csv'
    if not os.path.exists(out_file_name):
        df.to_csv(out_file_name, mode='w', header=True, index=False, encoding='utf-8')
    else:
        df.to_csv(out_file_name, mode='a', header=False, index=False, encoding='utf-8')
```

```python
filename = out_file_name
df = pd.read_csv(filename, encoding='utf8', dtype=str, escapechar='\\',error_bad_lines=False,engine='python')
```

```python
drop = df[pd.isnull(df['date']) == True]
```

```python
df = df.drop(drop.index)
```

```python
arr = ['date','anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust', 'neutral']
df = df.sort_values('date')
df.to_csv('./corona/april_sentiment_date_sorted.csv', mode='w', header=True, index=False, encoding='utf-8')
df
```

```python
df = pd.read_csv('./corona/april_sentiment_date_sorted.csv', encoding='utf8', dtype=str, escapechar='\\',error_bad_lines=False,engine='python')
df[['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust', 'neutral']] = df[['anger', 'anticipation', 'disgust', 'fear', 'joy', 'sadness', 'surprise', 'trust', 'neutral']].apply(pd.to_numeric)
df = df.groupby(['date']).sum()
df.to_csv('./corona/april_sentiment_date_sorted_merged.csv', mode='w', header=True, index=True, encoding='utf-8')
```
